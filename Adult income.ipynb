# 17. Replace income values ['<=50k','>=50k'] with 0 and 1 

```python
def income_data(inc):
    if inc=='<=50k':
      return 0
    else:
        return 1
        
```

```python
data['income']=data['income'].apply(income_data)
```

```python
data.head(10)
```

```python
data.replace(to_replace=['<=50k','>=50k'],value=[0,1],inplace=True)
```
